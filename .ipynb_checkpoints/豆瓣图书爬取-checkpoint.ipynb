```python
import requests as re 
from bs4 import BeautifulSoup  as bs
import pandas as pd
from openpyxl import Workbook
```

```python
book_data = pd.DataFrame()
b1=pd.DataFrame()
#取出urls列表
urls = ['https://book.douban.com/tag/%E5%93%B2%E5%AD%A6?start={}&type=T'.format(str(i)) for i in range(0,200,20)]
#声明一个存储书名的list
book_list = []
#声明一个存储评分的list
num_list = []
#声明一个字典
book_dict={}
def get_titles (urls):
    global book_data
    global book_list
    global num_list
    global book_dict
    page=re.get(urls)
    bf_page=bs(page.content,'html.parser')
    #爬取书名
    name =bf_page.find_all('h2')
    #爬取评分
    num =bf_page.find_all(class_=('rating_nums'))
    a = 0
    for i in name:
        book_name = i.text
        t = book_name.split()
        book_name3 = "".join(t) 
        book_list.append(book_name3)
        a=a+1
        if a>19:
            break
    for i in num: 
        book_num=i.text
        t=book_num.split()
        book_num2="".join(t) 
        num_list.append(book_num2)
  #字典添加
    book_dict={'书名':book_list,'评分':num_list}
    #book_data = pd.DataFrame(book_dict) 

for i in urls:
    get_titles(i)
#get_titles('https://book.douban.com/tag/%E5%93%B2%E5%AD%A6?start=180&type=T')
#book_data['评分'] = book_data['评分'].astype('float64')
bc={book_list}
bl=pd.DataFrame(bc)
print(b1)
"""
wb = Workbook()
ws = wb.active 
ws['A1'] = [1,2,3] 
ws['B1'] = 2
wb.save("/Users/wangjian/Desktop/test.xlsx")
"""
```
